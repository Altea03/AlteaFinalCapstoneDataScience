Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Definir la función de color
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'  # Éxito
    else:
        return 'red'    # Fallo

# 4. Crear la columna de color en el dataframe
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
```

```python
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# --- PASO 3: Añadir los marcadores al Cluster ---
for index, record in spacex_df.iterrows():
    # Crear el marcador individual
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        # Esto crea el pin blanco con el interior verde o rojo (como en tu foto)
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=record['Launch Site']
    )
    
    # Añadir el marcador al grupo (cluster)
    marker_cluster.add_child(marker)

# --- PASO 4: Mostrar el mapa ---
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# 1. Coordenadas del sitio de lanzamiento (CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# 2. Coordenadas de la costa más cercana (Las que pusiste en el ejemplo)
coastline_lat = 28.56367
coastline_lon = -80.57163

# 3. Calcular la distancia con tu función
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

print("Distancia a la costa:", distance_coastline, "km")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

# 1. Definir coordenadas (Sitio de lanzamiento y Costa)
launch_site_lat = 28.563197
launch_site_lon = -80.576820
coastline_lat = 28.56367
coastline_lon = -80.57163

# 2. Calcular la distancia
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# 3. Crear el marcador de distancia (Distance Marker)
distance_marker = folium.Marker(
   [coastline_lat, coastline_lon],
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
       )
   )

# 4. Añadir el marcador al mapa
site_map.add_child(distance_marker)

# 5. Dibujar la línea entre los dos puntos (PolyLine)
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)

# Mostrar el mapa
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

# 1. Definir coordenadas del sitio de lanzamiento (CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# 2. Definir las coordenadas de los puntos de interés cercanos (Ejemplos reales)
# [Latitud, Longitud]
closest_highway = [28.56335, -80.57085]   # Samuel C Phillips Pkwy
closest_railroad = [28.57206, -80.58525]  # NASA Railroad
closest_city = [28.10473, -80.64531]      # Melbourne (Ciudad cercana)

# Lista para iterar
locations = [
    (closest_highway, 'Highway'),
    (closest_railroad, 'Railway'),
    (closest_city, 'City')
]

# 3. Bucle para dibujar las 3 líneas y marcadores
for coordinate, label in locations:
    # Calcular distancia
    distance = calculate_distance(launch_site_lat, launch_site_lon, coordinate[0], coordinate[1])
    
    # Crear el Marcador de Distancia (Texto)
    distance_marker = folium.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s: %s KM</b></div>' % (label, "{:10.2f}".format(distance)),
        )
    )
    
    # Crear la Línea
    line = folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], coordinate],
        weight=1
    )
    
    # Añadir al mapa
    site_map.add_child(distance_marker)
    site_map.add_child(line)

# Mostrar el mapa final
site_map
```

After you plot distance lines to the proximities, you can answer the following questions easily:

Are launch sites in close proximity to railways? Yes. Railways are needed to transport heavy rocket components that are too large for standard roads.

Are launch sites in close proximity to highways? Yes. Highways are essential for transporting personnel, light equipment, and daily supplies.

Are launch sites in close proximity to coastline? Yes. This is critical for safety; if a launch fails, debris falls into the ocean rather than on land.

Do launch sites keep certain distance away from cities? Yes. Launch sites are located far from heavily populated areas to minimize the risk to human life and property in case of an accident.
